```python
dataset = pd.read_csv("plisbisa.csv")
dataset.head(13)
```

DISTRIBUSI DATA PADA PURCHASE PRICE, CURRENT PRICE, DAN GAIN/LOSS

```python
import pandas as pd
import matplotlib.pyplot as plt


# Replace non-numeric characters and convert to float
dataset['Purchase Price'] = dataset['Purchase Price'].replace('[\$,]', '', regex=True).astype(float)
dataset['Current Price'] = dataset['Current Price'].replace('[\$,]', '', regex=True).astype(float)
dataset['Gain/Loss'] = dataset['Gain/Loss'].replace('[\$,]', '', regex=True).astype(float)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].hist(dataset['Purchase Price'], bins=10, color='yellow', edgecolor='black')
axes[0].set_title('Distribution - Purchase Price')
axes[0].set_xlabel('Purchase Price')
axes[0].set_ylabel('Frequency')

axes[1].hist(dataset['Current Price'], bins=10, color='blue', edgecolor='black')
axes[1].set_title('Distribution - Current Price')
axes[1].set_xlabel('Current Price')
axes[1].set_ylabel('Frequency')

axes[2].hist(dataset['Gain/Loss'], bins=10, color='gray', edgecolor='black')
axes[2].set_title('Distribution - Gain/Loss')
axes[2].set_xlabel('Gain/Loss')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



top_dividend = dataset.sort_values(by='Dividend/share', ascending=False).head(5)

plt.figure(figsize=(12, 6))
sns.barplot(data=top_dividend, x='Stock Name', y='Dividend/share', hue='Stock Name', palette='viridis', dodge=False)
plt.title('5 Stock Name (symbol) dengan Dividend/Share Tertinggi')
plt.xlabel('Stock Name')
plt.ylabel('Dividend/Share')
plt.show()
```

```python
top_market_value = dataset.sort_values(by='Market Value').head(5)

top_market_value['Stock Label'] = top_market_value['Stock Name'] + ' (' + top_market_value['Symbol'] + ')'

plt.figure(figsize=(10, 6))
plt.bar(top_market_value['Stock Label'], top_market_value['Market Value'])
plt.xlabel('Stock Name')
plt.ylabel('Market Value')
plt.title('5 Stock Name (symbol) dengan Market Value Tertinggi')
plt.tight_layout()
plt.show()
```

```python
top_Annual_Yield = dataset.sort_values(by='Annual Yield', ascending=False).head(7)

top_Annual_Yield['Stock Label'] = top_Annual_Yield['Stock Name'] + ' (' + top_Annual_Yield['Symbol'] + ')'

plt.figure(figsize=(15, 6))
plt.bar(top_market_value['Stock Label'], top_market_value['Annual Yield'], color='skyblue')
plt.xlabel('Stocks')
plt.ylabel('Annual Yield (%)')
plt.title('7 Stocks Name(symbol) dengan Annual Yield tertinggi')
plt.tight_layout()
plt.show()
```
